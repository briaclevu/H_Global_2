Notebook to explore and visualize change in H<sup>+</sup> amplitude (&Delta;H<sup>+</sup>) in CMIP5 scenarios using PyFerret
---
Following the study of LE VU et al, in prep. about the acidification in Mediterranean Sea lets have a look on the amplitude change - if any - in the CMIP5 ptojection for 2100. Start with IPSL.

#### First thing is to install the ferret extension

```python
%install_ext https://raw.github.com/PBrockmann/ipython-ferretmagic/master/ferretmagic.py
```

#### Then load the extension

```python
%load_ext ferretmagic
```

#### Build a list of files from http://ocmip5.ipsl.fr/FileFinderAR5/ in remote access and open it (Grid IPSL and Output)

```python
listfiles = (
"http://dods.ipsl.jussieu.fr/cgi-bin/nph-dods/ocmip/phase5/OUTPUT/GRID/IPSL_IPSL-CM4_grid.nc" ,
"http://vesg3.ipsl.fr/thredds/dodsC/esg_data/CMIP5/merge/IPSL/IPSL-CM5A-LR/rcp85/mon/ocnBgchem/Omon/r1i1p1/latest/ph/ph_Omon_IPSL-CM5A-LR_rcp85_r1i1p1_200601-210512.nc" ,
"http://vesg3.ipsl.fr/thredds/dodsC/esg_data/CMIP5/merge/IPSL/IPSL-CM5A-LR/rcp85/mon/ocnBgchem/Omon/r1i1p1/latest/ph/ph_Omon_IPSL-CM5A-LR_rcp85_r1i1p1_210601-220512.nc" ,
"http://vesg3.ipsl.fr/thredds/dodsC/esg_data/CMIP5/merge/IPSL/IPSL-CM5A-LR/rcp85/mon/ocnBgchem/Omon/r1i1p1/latest/ph/ph_Omon_IPSL-CM5A-LR_rcp85_r1i1p1_220601-230012.nc"
)
for f in listfiles:
    print f
    %ferret_run 'use "%s"' % f
```

#### Now use extension cell (%%ferret) or line (%ferret_run) to execute commande

```python
%%ferret
show grid PH
sh d
```

#### Define H<sup>+</sup> and &Delta;H<sup>+</sup> present and futur

```python
%%ferret
let H1=10^(-1*PH[d=2,l=1:12])*1e9
let H2=10^(-1*PH[d=3,l=1:12])*1e9
let D_H_2006=H1[l=@max]-H1[l=@min]
let D_H_2106=H2[l=@max]-H2[l=@min]
```

#### Plot surface  delta amplitude of H<sup>+</sup> in 2006 and 2106

```python
%ferret_run -q -s 500,300 'shade/levels=(0,10,0.5) D_H_2006' % locals()
%ferret_run -q -s 500,300 'shade/levels=(0,10,0.5) D_H_2106' % locals()
```

#### Plot ratio between 2106 and 2006 amplitude (<1 it's a decrease) and relative change (<0 it's a decrease)

```python
%ferret_run -q -s 500,300 'shade/levels=(-1,5,1) D_H_2106/D_H_2006' % locals()
%ferret_run -q -s 500,300 'shade/levels=(-1,5,1) (D_H_2106-D_H_2006)/D_H_2006' % locals()
```

First conclusion
--
* There is general increase in amplitude between 2006 and 2100 similar to the one in the Mediterranean Sea (a doubling)
* We observe in high lattitude and frontal area (South Iceland, Southe Africa, South Indian and Arctic) a important increase (5 fold the 2006 amplitude)
* Sometimes there is a decrease (south Pacific, Antarctic,...) but seema adjacent to an important increase area which could be due the interannual variability

Perspective
--
* The same study with decadal monthly climatology is needed to smooth the interannual variability
